```python
import numpy as np
import cv2 as cv
# Create a black image
img = np.zeros((512,512,3), np.uint8)
# Draw a diagonal blue line with thickness of 5 px
cv.line(img,(0,0),(511,511),(255,0,0),5)
```

```python
cv.rectangle(img,(384,0),(510,128),(0,255,0),3)
```

```python
cv.rectangle(img,(384,0),(510,128),(0,255,0),3)
```

```python
cv.circle(img,(447,63), 63, (0,0,255), -1) # center radius color thikniess(A negative value indicates that the circle should be filled completely.)
```

```python
cv.ellipse(img,(256,256),(100,50),0,0,180,255,-1) # center (max,min)radius angle start_angle end_angle color thickness
```

```python
pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts = pts.reshape((-1,1,2))
cv.polylines(img,[pts],True,(0,255,255))
```

```python
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(img,'OpenCV',(10,500), font, 4,(255,255,255),2,cv.LINE_AA)
```
